```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with CSE-B.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
data
```

```python
pattern =r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s\w+\s-\s'
```

```python
message=re.split(pattern,data)[1:]
message
```

```python
dates=re.findall(pattern,data)
dates
```

```python
# Regular expression pattern to capture date, time and am/pm
pattern = re.compile(r'(\d{2}/\d{2}/\d{4}), (\d{1,2}:\d{2}\u202f[ap]m)')

# Extracting the matches
matches = [pattern.search(line) for line in dates]
dates_times = [(match.group(1), match.group(2)) for match in matches if match]

# Create a DataFrame
df = pd.DataFrame(dates_times, columns=['Date', 'Time'])
print(df)
```

```python
date=[]
times=[]
for i in dates:
    date.append(i.split(", ")[0])
    times.append(i.split(", ")[1])
time=[]
for i in times:
    time.append(i.split("-")[0])
df=pd.DataFrame({
    'user_message':message,
    'date':date,
    'time':time
})
df
```

```python


# Initialize empty lists for users and messages
users = []
messages = []

# Iterating over each message in 'message' column (assuming this is where messages are)
for i in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', i)
    if entry[1:]:  # user name is present
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0].strip())

# Adding 'user' and 'message' columns to the DataFrame
df['user'] = users
df['user_message'] = messages

# Displaying the final DataFrame
df
```

```python
#df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month_name()
```

```python
df
```

```python
from datetime import datetime
# Conversion function with error handling
def convert_to_24hr(time_str):
    try:
        # Remove narrow no-break space and strip any extra spaces
        time_str_cleaned = time_str.replace('\u202f', '').strip()
        # Convert to 24-hour format
        return datetime.strptime(time_str_cleaned, '%I:%M%p').strftime('%H:%M')
    except ValueError:
        return 'Invalid Format'

# Applying the conversion
df['Time_24hr'] = df['time'].apply(convert_to_24hr)

print(df)
```
